# Soal 4:

```python
print('Saat ini, jumlah usia Andi & Budi = 49 Tahun')
print('Rasio usia Andi & Budi = 0.4')
print('Berapakah usia Andi & Budi 2 Tahun lagi?\n')


totalUsia = 49
rasioUsia = 0.4

# asumsi Budi yang lebih tua
# usiaAndi / usiaBudi = 0.4
# usiaAndi = 0.4 x usiaBudi
# usiaBudi + usiaAndi = usiaBudi + (0.4 usiaBudi) = 49
# 1.4 x usiaBudi = 49
# usiaBudi = 49/1.4

usiaBudi = int(totalUsia / (1 + rasioUsia))
usiaAndi = int(totalUsia - usiaBudi)

print(f'Usia Andi sekarang = {usiaAndi}')
print(f'Usia Budi sekarang = {usiaBudi}')
print()
print('2 Tahun lagi:')
print(f'Usia Andi 2 Tahun lagi = {usiaAndi + 2}')
print(f'Usia Budi 2 Tahun lagi = {usiaBudi + 2}')
```

# Soal 5:

```python
import math

print('Jarak Mobil A & B = 120 Km.')
print('A berjalan 60 Km/h dari timur.')
print('B berjalan 40 Km/h dari barat.')
print('A & B start pukul 9 WIB.\n')
print('Jam berapa A & B bertabrakan?\n')

jarak = 120
vA = 60
vB = 40
jamStart = 9

vTotal = vA + vB

ETATabrakan = (jarak / vTotal)

jamTabrakan = jamStart + math.floor(ETATabrakan)
menitTabrakan = math.floor(ETATabrakan * 60) % 60

print(f'A & B akan bertabrakan pada jam {jamTabrakan}.{menitTabrakan} WIB')



```

# Soal 6:

```python
hargaApel = 10_000
hargaJeruk = 15_000
hargaAnggur = 20_000

qtyApel = int(input('Masukkan Jumlah Apel: '))
print(f'Jumlah Apel = {qtyApel}')

qtyJeruk = int(input('Masukkan Jumlah Jeruk: '))
print(f'Jumlah Jeruk = {qtyJeruk}')

qtyAnggur = int(input('Masukkan Jumlah Anggur: '))
print(f'Jumlah Anggur = {qtyAnggur}')

totalApel = qtyApel * hargaApel
totalJeruk = qtyJeruk * hargaJeruk
totalAnggur = qtyAnggur * hargaAnggur
totalSemua = totalApel + totalJeruk + totalAnggur

print('\nDetail Belanja\n')

print(f'Apel : {qtyApel} x {hargaApel} = {totalApel}')
print(f'Jeruk : {qtyJeruk} x {hargaJeruk} = {totalJeruk}')
print(f'Anggur : {qtyAnggur} x {hargaAnggur} = {totalAnggur}')

print(f'\nTotal : {totalSemua}')
```
