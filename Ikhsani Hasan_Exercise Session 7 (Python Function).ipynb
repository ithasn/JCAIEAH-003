```python
def ganjilGenap(listAngka):
    listAngka = list(map(lambda angka: 'Ganjil' if angka%2 == 1 else 'Genap', listAngka))
    return listAngka
```

```python
print(f'{list1} = {ganjilGenap(list1)}')
print(f'{list2} = {ganjilGenap(list2)}')
print(f'{list3} = {ganjilGenap(list3)}')
print(f'{list4} = {ganjilGenap(list4)}')
```

```python
listBuahSeller = [
    {'nama' : 'Apel', 'qty' : 20, 'harga' : 10_000},
    {'nama' : 'Jeruk', 'qty' : 15, 'harga' : 15_000},
    {'nama' : 'Anggur', 'qty' : 25, 'harga' : 20_000}
]
```

```python
def integerInput(text):
    while True:
        userInput = input(text)
        if userInput.isdigit():
            userInput = int(userInput)
            break
        else:
            print('Masukkan bilangan bulat\n')
            continue
    
    return userInput
```

```python
def tampilkanBuah(listBuah):
    print('Index\tNama\t\tQty\tHarga')
    print('----------------------------------------')
    for i, item in enumerate(listBuah):
        print(f"{i}\t{item['nama']}\t\t{item['qty']}\t{item['harga']}")
    print()
```

```python
def tambahBuah(listBuah):
    inputNama = input('Masukkan nama buah: ')
    inputQty = integerInput('Masukkan stock buah: ')
    inputHarga = integerInput('Masukkan harga buah: ')

    buahBaru = {'nama': inputNama, 'qty' : inputQty, 'harga' : inputHarga}

    listBuah.append(buahBaru)

    return listBuah
```

```python
def hapusBuah(listBuah):
    while True:
        inputIndex = integerInput('Masukkan index buah: ')

        if inputIndex <= len(listBuah)-1:
            del listBuah[inputIndex]
            break
        else:
            print('Index tersebut tidak ada\n')
            continue
    
    return listBuah
```

```python
def beliBuah(listBuah, cart):
    beliLagi = 'y'
    
    while beliLagi == 'y':
        while True:
            print('Stock buah:')
            tampilkanBuah(listBuah)
            inputIndex = integerInput('Masukkan index buah: ')

            if inputIndex <= len(listBuah)-1:
                inputQty = integerInput('Masukkan jumlah buah yang ingin dibeli: ')
                if inputQty > listBuah[inputIndex]['qty']:
                    print(f"Stock tidak cukup, stock {listBuah[inputIndex]['nama']} tinggal {listBuah[inputIndex]['qty']}\n")
                    print('Isi cart:')
                    tampilkanBuah(cart)
                    break
                else:
                    indexExisting = None

                    for i, item in enumerate(cart):
                        if item['nama'] == listBuah[inputIndex]['nama']:
                            indexExisting = i
                            break

                    if indexExisting is not None:
                        cart[indexExisting]['qty'] += inputQty
                        cart[indexExisting]['harga'] += listBuah[inputIndex]['harga']*inputQty
                    else:
                        cart.append({'nama':listBuah[inputIndex]['nama'],'qty':inputQty, 'harga':listBuah[inputIndex]['harga']*inputQty})

                    listBuah[inputIndex]['qty'] -= inputQty

                    print('Isi cart:')
                    tampilkanBuah(cart)
                    break
            else:
                print('Index tersebut tidak ada\n')
                continue
    
        beliLagi = input('Mau beli yang lain? (y/n)').lower()

    return listBuah, cart
```

```python
def bayar(cart):
    total = sum(item['harga'] for item in cart)
    print(f'Total yang harus dibayar = {total}\n')
    while True:
        jumlahBayar = integerInput('Masukkan jumlah uang: ')
        print(f'Jumlah uang: {jumlahBayar}\n')

        selisihBayar = jumlahBayar - total

        if selisihBayar < 0:
            print('Transaksi anda dibatalkan')
            print(f'Uangnya kurang sebesar {-selisihBayar}\n')
            continue
        elif selisihBayar == 0:
            print('Terima kasih\n')
            break
        else:
            print('Terima kasih')
            print(f'Uang kembali anda: {selisihBayar}\n')
            break
```

```python
def mainMenu(listBuah):

    cartUser = []

    print('Selamat Datang di Pasar Buah\n')
    print('List Menu:')
    print('1. Menampilkan Daftar Buah')
    print('2. Menambah Buah')
    print('3. Menghapus Buah')
    print('4. Membeli Buah')
    print('5. Exit Program')
    print()

    while True:
        inputMenu = input('Masukkan angka menu yang ingin dijalankan: \n')

        if inputMenu == '1':
            print('Stock buah:')
            tampilkanBuah(listBuah)
        elif inputMenu == '2':
            tambahBuah(listBuah)
        elif inputMenu == '3':
            hapusBuah(listBuah)
        elif inputMenu == '4':
            listBuahSeller, cartUser = beliBuah(listBuah, cartUser)
            bayar(cartUser)
            cartUser = []
        elif inputMenu == '5':
            break
        else:
            print('Masukkan hanya 1-5\n')
```
