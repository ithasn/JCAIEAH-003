```python
import numpy as np
import pandas as pd
```

```python
def intInput(text):
    while True:
        try:
            x = int(input(text))
            break
        except ValueError:
            print('Masukkan bilangan integer.\n')
            continue
    return x
```

---
# Soal 1

```python
startArr = intInput('Masukkan nilai Start: ')
stopArr = intInput('Masukkan nilai Stop: ')

print(f'Nilai Start = {startArr}')
print(f'Nilai Stop {stopArr}')
```

#### Array 1D

```python
arr = np.arange(startArr, stopArr)
arr
```

### Array 2D

```python
arr.reshape(3,9)
```

### Array 3D

```python
arr.reshape(3,3,3)
```
