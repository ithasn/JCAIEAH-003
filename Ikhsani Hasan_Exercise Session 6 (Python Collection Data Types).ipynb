# Soal 3:

```python
listBuah = [
    {'nama' : 'Apel', 'qty' : 20, 'harga' : 10_000},
    {'nama' : 'Jeruk', 'qty' : 15, 'harga' : 15_000},
    {'nama' : 'Anggur', 'qty' : 25, 'harga' : 20_000}
]

cart = []

#____________________________________________________________

print('Selamat Datang di Pasar Buah\n')

print('List Menu:')
print('1. Menampilkan Daftar Buah')
print('2. Menambah Buah')
print('3. Menghapus Buah')
print('4. Membeli Buah')
print('5. Exit Program')
print()

while True:
    inputMenu = input('Masukkan angka menu yang ingin dijalankan: \n')
    
    # Menu 1: Menampilkan Daftar Buah
    if inputMenu == '1':
        print('Index\tNama\t\tStock\tHarga')
        print('----------------------------------------')
        for i, item in enumerate(listBuah):
            print(f'{i}\t{item['nama']}\t\t{item['qty']}\t{item['harga']}')
        print()

    # Menu 2: Menambah Buah    
    elif inputMenu == '2':
        inputNama = input('Masukkan nama buah: ')

        while True:
            inputQty = input('Masukkan stock buah: ')
            if inputQty.isdigit():
                inputQty = int(inputQty)
                break
            else:
                print('Masukkan bilangan bulat\n')
                continue

        while True:
            inputHarga = input('Masukkan harga buah: ')
            if inputHarga.isdigit():
                inputHarga = int(inputHarga)
                break
            else:
                print('Masukkan bilangan bulat\n')
                continue

        buahBaru = {'nama': inputNama, 'qty' : inputQty, 'harga' : inputHarga}

        listBuah.append(buahBaru)

    # Menu 3: Menghapus Buah
    elif inputMenu == '3':
        while True:
            inputIndex = input('Masukkan index buah yang ingin dihapus: ')
            if inputIndex.isdigit():
                inputIndex = int(inputIndex)
                if inputIndex <= len(listBuah)-1:
                    del listBuah[inputIndex]
                    break
                else:
                    print('Index tersebut tidak ada\n')
                    continue
            else:
                print('Masukkan hanya bilangan bulat\n')
                continue
        
    # Menu 4: Membeli Buah
    elif inputMenu == '4':
        while True:
            inputIndex = input('Masukkan index buah yang ingin dibeli: ')
            if inputIndex.isdigit():
                inputIndex = int(inputIndex)
                if inputIndex <= len(listBuah)-1:
                    
                    while True:
                        inputQty = input('Masukkan jumlah buah yang ingin dibeli: ')
                        if inputQty.isdigit():
                            inputQty = int(inputQty)
                            if inputQty > listBuah[inputIndex]['qty']:
                                print(f'Stock tidak cukup, stock {listBuah[inputIndex]['nama']} tinggal {listBuah[inputIndex]['qty']}')
                                print('Isi cart:')
                                print('Nama\tQty\tHarga')
                                print('----------------------------------------')
                                for i in cart:
                                    print(f'{i['nama']}\t{i['qty']}\t{i['harga']}')
                                print()
                                break
                                
                            else:
                                cart.append({'nama':listBuah[inputIndex]['nama'],'qty':inputQty, 'harga':listBuah[inputIndex]['harga']*inputQty})
                                listBuah[inputIndex]['qty'] -= inputQty
                                totalBelanja = sum(itemBelanja['harga'] for itemBelanja in cart)
                                print('Isi cart:')
                                print('Nama\tQty\tHarga')
                                print('----------------------------------------')
                                for i in cart:
                                    print(f'{i['nama']}\t{i['qty']}\t{i['harga']}')
                                print()
                                break
                                
                        else:
                            print('Masukkan hanya bilangan bulat\n')
                            continue
                    
                    inputBeliLagi = input('Mau beli yang lain? (y/n)')
                    if inputBeliLagi == 'y' or inputBeliLagi == 'Y':
                        continue
                    else:
                        print(f'Total yang harus dibayar = {totalBelanja}\n')
                        while True:
                            jumlahBayar = input('Masukkan jumlah uang: ')
                            if jumlahBayar.isdigit():
                                jumlahBayar = int(jumlahBayar)

                                print(f'Jumlah uang: {jumlahBayar}\n')

                                selisihBayar = jumlahBayar - totalBelanja

                                if selisihBayar < 0:
                                    print('Transaksi anda dibatalkan')
                                    print(f'Uangnya kurang sebesar {-selisihBayar}\n')
                                    continue
                                elif selisihBayar == 0:
                                    print('Terima kasih')
                                    break
                                else:
                                    print('Terima kasih')
                                    print(f'Uang kembali anda: {selisihBayar}')
                                    break
                            else:
                                print('Masukkan bilangan bulat\n')
                                continue

                        break    

                else:
                    print('Index tersebut tidak ada\n')
                    continue
            else:
                print('Masukkan hanya bilangan bulat\n')
                continue

    # Menu 5: Exit Program
    elif inputMenu == '5':
        break

    else:
        print('Masukkan hanya 1-5\n')
    
```

# Soal 4:

```python
setFilmSaya = {'Avengers', 'Hulk', 'Iron Man 3', 'Wonder Woman', 'Batman'}
setFilmTeman = {'Avengers', 'Hulk', 'Iron Man 2', 'Superman', 'Wonder Woman'}
kesamaan = 0

print(setFilmSaya)
print(setFilmTeman)

for filmSaya in setFilmSaya:
    for filmTeman in setFilmTeman:
        if filmSaya == filmTeman:
            kesamaan += 1
        else:
            continue

print(f'Kesukaan Film kalian yang sama sebesar {(kesamaan*2)/(len(setFilmSaya)+len(setFilmTeman))*100}%')
```
