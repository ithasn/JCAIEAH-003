##### Bentuk 1:

```python
import math

baris = 9
mid = baris/2

for i in range(0,baris):
    if i < mid:
        jumlahBintang = math.ceil(mid - i)
    else:
        jumlahBintang = math.ceil((i+1) - mid)
    
    print('* '*jumlahBintang)
```

```python
import math

baris = 9
mid = baris/2

for i in range(0,baris):
    if i < mid:
        jumlahSpasi = math.ceil((mid-1) - i)
        jumlahBintang = (i*2) + 1
    else:
        jumlahSpasi = math.ceil((i+1) - (mid+1))
        jumlahBintang = ((baris-i)*2)-1
    
    print(('  '*jumlahSpasi)+('* '*jumlahBintang))
        
```

# Soal 2:

```python
stockApel = 5
stockJeruk = 7
stockAnggur = 6

hargaApel = 10_000
hargaJeruk = 15_000
hargaAnggur = 20_000

while True:
    qtyApel = input('Masukkan jumlah Apel" ')
    if qtyApel.isdigit():
        qtyApel = int(qtyApel)
        if qtyApel > stockApel:
            print('Jumlah yang dimasukkan terlalu banyak')
            print(f'Stock Apel tinggal : {stockApel}\n')
            continue
        else:
            print(f'Jumlah Apel: {qtyApel}\n')
            break
    else:
        print('Masukkan bilangan bulat\n')
        continue

while True:
    qtyJeruk = input('Masukkan jumlah Jeruk" ')
    if qtyJeruk.isdigit():
        qtyJeruk = int(qtyJeruk)
        if qtyJeruk > stockJeruk:
            print('Jumlah yang dimasukkan terlalu banyak')
            print(f'Stock Jeruk tinggal : {stockJeruk}\n')
            continue
        else:
            print(f'Jumlah Jeruk: {qtyJeruk}\n')
            break
    else:
        print('Masukkan bilangan bulat\n')
        continue

while True:
    qtyAnggur = input('Masukkan jumlah Anggur" ')
    if qtyAnggur.isdigit():
        qtyAnggur = int(qtyAnggur)
        if qtyAnggur > stockAnggur:
            print('Jumlah yang dimasukkan terlalu banyak')
            print(f'Stock Anggur tinggal : {stockAnggur}\n')
            continue
        else:
            print(f'Jumlah Anggur: {qtyAnggur}\n')
            break
    else:
        print('Masukkan bilangan bulat\n')
        continue

totalApel = qtyApel * hargaApel
totalJeruk = qtyJeruk * hargaJeruk
totalAnggur = qtyAnggur * hargaAnggur
totalSemua = totalApel + totalJeruk + totalAnggur

print('\nDetail Belanja\n')

print(f'Apel : {qtyApel} x {hargaApel} = {totalApel}')
print(f'Jeruk : {qtyJeruk} x {hargaJeruk} = {totalJeruk}')
print(f'Anggur : {qtyAnggur} x {hargaAnggur} = {totalAnggur}')

print(f'\nTotal : {totalSemua}\n')

while True:
    jumlahBayar = input('Masukkan jumlah uang: ')
    if jumlahBayar.isdigit():
        jumlahBayar = int(jumlahBayar)

        print(f'Jumlah uang: {jumlahBayar}\n')

        selisihBayar = jumlahBayar - totalSemua

        if selisihBayar < 0:
            print('Transaksi anda dibatalkan')
            print(f'Uangnya kurang sebesar {-selisihBayar}\n')
            continue
        elif selisihBayar == 0:
            print('Terima kasih')
            break
        else:
            print('Terima kasih')
            print(f'Uang kembali anda: {selisihBayar}')
            break
    else:
        print('Masukkan bilangan bulat\n')
        continue
```
